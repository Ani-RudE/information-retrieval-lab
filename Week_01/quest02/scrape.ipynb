```python
import os
import wikipediaapi

def download_wikipedia_articles(directory, titles):
    """Download Wikipedia articles and save them as text documents."""
    if not os.path.exists(directory):
        os.makedirs(directory)

    # Define the user agent
    user_agent = "scrape/1.0 (ani.dewangan2002@gmail.com)"
    
    wiki_wiki = wikipediaapi.Wikipedia('en', headers={'User-Agent': user_agent})

    for i, title in enumerate(titles, start=1):
        page = wiki_wiki.page(title)
        if page.exists():
            file_path = os.path.join(directory, f'wikipedia_article_{i}.txt')
            with open(file_path, 'w', encoding='utf-8') as file:
                file.write(page.text)
            print(f"Downloaded: {title}")
        else:
            print(f"Page '{title}' does not exist on Wikipedia.")
    
    print(f'{len(titles)} Wikipedia articles downloaded and saved in directory: {directory}')

# Set the directory path
directory_path = 'docs'

# List of 100 Wikipedia article titles
titles = [
    "Python (programming language)",
    "Natural language processing",
    "Machine learning",
    "Artificial intelligence",
    "Data science",
    "Big data",
    "Neural network",
    "Deep learning",
    "Computer vision",
    "Robotics",
    "Quantum computing",
    "Blockchain",
    "Cryptocurrency",
    "Internet of things",
    "Augmented reality",
    "Virtual reality",
    "Cybersecurity",
    "Cloud computing",
    "Software development",
    "Programming language",
    "Operating system",
    "Linux",
    "Windows (operating system)",
    "MacOS",
    "Open source",
    "GitHub",
    "Version control",
    "Database",
    "SQL",
    "NoSQL",
    "MongoDB",
    "Apache Hadoop",
    "Data warehouse",
    "Data mining",
    "Business intelligence",
    "Data visualization",
    "Computer graphics",
    "Game development",
    "Virtual machine",
    "Containerization",
    "Docker (software)",
    "Kubernetes",
    "Microservices",
    "DevOps",
    "Agile software development",
    "Scrum (software development)",
    "Kanban",
    "Lean software development",
    "Extreme programming",
    "Software testing",
    "Test-driven development",
    "Continuous integration",
    "Continuous deployment",
    "Software architecture",
    "Design patterns",
    "Software engineering",
    "Systems engineering",
    "Embedded system",
    "Real-time computing",
    "Internet protocol",
    "HTTP",
    "HTTPS",
    "World Wide Web",
    "HTML",
    "CSS",
    "JavaScript",
    "TypeScript",
    "React (JavaScript library)",
    "Vue.js",
    "Angular (web framework)",
    "Node.js",
    "Express.js",
    "REST (representational state transfer)",
    "GraphQL",
    "APIs",
    "JSON",
    "XML",
    "Web development",
    "Responsive web design",
    "Progressive web application",
    "SEO",
    "Digital marketing",
    "E-commerce",
    "Online advertising",
    "Social media",
    "Content management system",
    "WordPress",
    "Joomla",
    "Drupal",
    "Enterprise resource planning",
    "Customer relationship management",
    "Business process management",
    "Supply chain management",
    "Artificial general intelligence",
    "Superintelligence",
    "Transhumanism",
    "Singularity (technology)",
    "Philosophy of artificial intelligence",
    "Ethics of artificial intelligence",
    "History of artificial intelligence",
    "Future of artificial intelligence",
    "Automation",
    "Digital transformation",
    "Industry 4.0",
    "Smart cities",
    "Wearable technology"
]

# Download the Wikipedia articles
download_wikipedia_articles(directory_path, titles)
```
